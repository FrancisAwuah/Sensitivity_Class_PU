MLE

```python
import matplotlib.pyplot as plt

# Define the data
percentages = [20, 30, 40, 50]
f1_scores = {
    'PIG': [1, 1, 1, 1],
    'CAL': [1, 1, 1, 1],
    'MM': [1, 1, 1, 1],
}

# Set the style and size of the figure
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(8, 6))

# Plot the lines for each method
for method, scores in f1_scores.items():
    plt.plot(percentages, scores, label=method)

# Set the title and labels
plt.title('F1-Scores at different percentages')
plt.xlabel('Percentages')
plt.ylabel('F1-Scores')
plt.xticks(percentages)

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
