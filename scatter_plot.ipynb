```python
import matplotlib.pyplot as plt

# Define the data
class_prior_estimation_methods = ["MLE", "CV", "EI", "MBM"]
learning_methods = {
    'PIG': [0.5, 0.4, 0.4, 0.4],
    'CAL': [0.6, 0.4, 0.6, 0.6],
    'MM': [0.4, 0.4, 0.4, 0.4],
}


# Set the style and size of the figure
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(8, 6))

# Plot the scatter points for each method
for method, scores in learning_methods.items():
    plt.scatter(class_prior_estimation_methods, scores, label=method)

# Set the title and labels
plt.title('F1-Scores at different Class Prior Estimation Methods')
plt.xlabel('Class Prior Estimation Methods')
plt.ylabel('F1_scores')



# Add a legend
plt.legend()

# Show the plot
plt.show()
```
